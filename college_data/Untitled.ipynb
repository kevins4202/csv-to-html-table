```python
import pandas as pd

import os
```

```python
os.listdir()
```

```python
for d in os.listdir():
    if 'csv' in d:
        df = pd.read_csv(d)
        
        df.columns = ['Name','Affiliation','Homepage','Google Scholar ID','Citations','h-index','Earliest Paper']
        df.to_csv(d, index = False)
```
